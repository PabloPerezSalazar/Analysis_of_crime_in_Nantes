### Récupération et préparation des données

```python
# Get data

import os

current_folder = globals()['_dh'][0]

# Calculating path to the input data
data_location = os.path.join(current_folder,'data\crimes-et-delits-enregistres_PN.xlsx')

import pandas as pd
import openpyxl
import xlrd

multi_sheet_file = pd.ExcelFile(data_location, engine='openpyxl')
```

#### Créer un jeu de données pour chaque feuille du fichier Excel

```python
def dataframe_dict(multi_sheet_file):
    """
    Create one Dataframe for each sheet in a Excel file.
    """
    # Gets the sheet names in the file
    excel_sheet_names = multi_sheet_file.sheet_names

    dict_of_sheets = {}
    for sheet in excel_sheet_names:
        dict_of_sheets["sheet_" + sheet] = pd.read_excel(multi_sheet_file, sheet_name=sheet, header = 1)
         #  header = 1 for use second row as a name of columns
    
    return dict_of_sheets
```

```python
police_records_dict = dataframe_dict(multi_sheet_file)

# Names of dataframes
print(police_records_dict.keys())
```

```python
# Get one dataframe from dictionary
police_record_2012 = police_records_dict.get('sheet_Services PN 2012')

# View an extract from the dataframe
police_record_2012.head()
```

```python
# Iterating the columns
for column_name in police_record_2012.columns:
    if "nantes" in column_name.lower() : # Filter only records from Nantes
        print(column_name)
```

```python
police_record_2012 = police_record_2012[["SPAFA NANTES ATLANTIQUE","CIAT CENTRAL DE NANTES"]].apply(pd.to_numeric)

print(police_record_2012.info())
police_record_2012.describe()
```

####  Conter dans une seule colonne tous les crimes commis dans cette ville

```python
# Calculate in a single column all the crimes committed in that city
police_record_2012["Nantes"] = police_record_2012["SPAFA NANTES ATLANTIQUE"] + police_record_2012["CIAT CENTRAL DE NANTES"]
```

#### Organiser / concatener tout dans un seul jeu de données

```python
# Transforming the police records dictionary to list
police_records_list = []

for key, value in police_records_dict.items():
    police_records_list.append(value)
```

```python
police_records_list[1].head() # police record 2013
```

```python
police_records_new_list = []

for dataframe in police_records_list :
    
    # Filter only records from Nantes and change the datatype from text to numeric
    dataframe[["SPAFA NANTES ATLANTIQUE","CIAT CENTRAL DE NANTES"]] = dataframe[["SPAFA NANTES ATLANTIQUE","CIAT CENTRAL DE NANTES"]].apply(pd.to_numeric)
    # Save only one column from Nantes
    dataframe["Delits enregistres"] = dataframe["SPAFA NANTES ATLANTIQUE"] + dataframe["CIAT CENTRAL DE NANTES"]
    # Rename columns of crimes types
    dataframe = dataframe.rename(columns={"Libellé index \ CSP": "Libellé du delit", "ANNEE":"Annee"})
    # New dataframe
    dataframe = dataframe[["Libellé du delit", "Type de crime","Delits enregistres","Annee"]]
    police_records_new_list.append(dataframe)

police_records_new_list[1]
```

```python
# Concatenate all dataframes
police_records_copilation = pd.concat(police_records_new_list)
```

```python
police_records_copilation
```

```python
# Save new dataframe
path_data = os.path.join(current_folder,'data\police_records_nantes.csv')
police_records_copilation.to_csv(path_data) 
```


## 3. Si la tendance se poursuit, quels sont les crimes qui connaîtront la plus forte croissance à l'avenir.

### Méthode de prévision ARIMA

Une alternative (en l'absence de données préalables suffisantes relatives aux variables explicatives) est la prédiction basée sur une stratégie autorégressive comme le modèle ARIMA (autoregressive integrated moving average). Cette technique est implémentée notamment pour les prévisions des séries temporelles car des éléments tels qu’une possible saisonnalité y sont pris en compte, et le résultat est une extrapolation de l'avenir en fonction des éléments passés.

```python
import statsmodels.api as sm
import statsmodels.graphics.api as smg

# preparation of data for the model ML - Arima

vols_ari = vols
# datatime as index
vols_ari.index = vols_ari['Annee']
vols_ari = vols_ari['Delits enregistres']
vols_ari.head()
```

```python
# modélisation et entrainement
with warnings.catch_warnings(): #to Turn off Warnings
    warnings.simplefilter('ignore')
    mod = sm.tsa.statespace.SARIMAX(vols_ari,trend='c', order=(1,0,0)) # (p,d,q) for a non-seasonal ARIMA model

    results = mod.fit(disp=False)

# See the results of the forecast for two next years
print(results.forecast(steps=2))
```

```python
df_fcast = pd.DataFrame(results.forecast(steps=3)) # forecast for three next years
df_fcast["date"] = df_fcast.index
df_fcast = df_fcast.rename(columns={0: "delits"}).reset_index(drop=True)

print(df_fcast.info())
df_fcast
```

```python
base = alt.Chart(df_fcast).mark_bar(color="orange").encode(
    x='date:T',
    y='delits:Q'
)

vols_visualisation = alt.Chart(vols.reset_index(drop=True)).encode(
    alt.X("Annee", axis=alt.Axis(title="", grid=False)),
    alt.Y("Delits enregistres:Q", axis=alt.Axis(title="Nombre de delits enregistres", grid=False)),
    ).properties(
        title="Estimation de l'évolution de la criminalité dans les années à venir",
        width=500,
        height=300,
    )

base + base.transform_loess("date", "delits").mark_line(color =  'orange', size=0.5).encode().interactive() + vols_visualisation.mark_bar() + vols_visualisation.transform_loess("Annee", "Delits enregistres").mark_line(size=0.5).encode().interactive()
```

Si nous ne prenons en compte que les données historiques des enregistrements de ce délit, selon notre modèle de prévision, les prochaines années seront marquées par une augmentation des vols.

#### Prévisions pour chaque délit

```python
def altair_visualization_predictions(police_records_agregat):
    """
    This is where the altair chart is defined, using the pandas dataframe.

    :returns: altair chart
    
    https://altair-viz.github.io/user_guide/compound_charts.html
    """
    charts = []
    for delit in police_records_agregat['Type de crime'].unique() :        
        police_records_filtre = police_records_agregat[police_records_agregat['Type de crime'].str.contains(delit)]
        
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            police_records_filtre['Annee'] = police_records_filtre['Annee'].astype(str)\
                            .apply(lambda x: dt.datetime.strptime(x,'%Y'))

            police_records_ari = police_records_filtre
            # datatime as index
            police_records_ari.index = police_records_ari['Annee']
            police_records_ari = police_records_ari['Delits enregistres']

            # modélisation et entrainement
            mod = sm.tsa.statespace.SARIMAX(police_records_ari,trend='c', order=(1,0,0)) # (p,d,q) pour un modèle ARIMA non saisonnier

            results = mod.fit(disp=False)

            df_fcast = pd.DataFrame(results.forecast(steps=3))
            df_fcast["date"] = df_fcast.index
            df_fcast = df_fcast.rename(columns={0: "delits"}).reset_index(drop=True)

            prediction = alt.Chart(df_fcast).mark_bar(color="orange").encode(
            x='date:T',
            y='delits:Q'
            )

            visualisation = alt.Chart(police_records_filtre.reset_index(drop=True)).encode(
                alt.X("Annee", axis=alt.Axis(title="", grid=False)),
                alt.Y("Delits enregistres:Q", axis=alt.Axis(title="Nombre de delits enregistres", grid=False)),
                ).properties(
                    title="Estimation de l'évolution des " + delit + " dans les années à venir",
                    width=500,
                    height=300,
                )
            
            charts.append(alt.layer(
                prediction, 
                prediction.transform_loess("date", "delits").mark_line(color =  'orange', size=0.5).encode().interactive(), 
                visualisation.mark_bar(),
                visualisation.transform_loess("Annee", "Delits enregistres").mark_line(size=0.5).encode().interactive()
            ))
            
    return charts
```

```python
for graphique in altair_visualization_predictions(police_records_agregat):
    graphique.display()
```

Selon l'analyse visuelle de ces projections, les crimes qui risquent d'augmenter dans les années à venir sont les suivants : 

- Atteintes à l'environnement
- Harcèlement non sexuel
- Infractions au droit de construction
- Infractions aux conditions d'entrée des étrangers
- Proxénétisme

## 4. Est-ce qu'il a une correlation entre le crime et le chomage ?

### Récupération et préparation des données

```python
# Get data
taux_chomage = pd.read_csv(r'data\taux_chomage_pays_de_la_loire.csv', sep=';')
```

```python
print(taux_chomage.info())
taux_chomage.head()
```

```python
taux_chomage['Zone'].unique()
```

```python
# Transform the data type from years to date
taux_chomage['Date'] = taux_chomage['Date'].apply(pd.to_datetime)
taux_chomage['Annee'] = taux_chomage['Date'].dt.year.astype('Int64')

# Filter the Nantes area
taux_chomage = taux_chomage[(taux_chomage['Zone'] == 'Nantes')]

taux_chomage.head()
```

#### Calculer un taux de chômage annuel moyen

```python
# Calculate an average annual unemployment rate
taux_chomage = taux_chomage.groupby(['Annee'],as_index=False)['Taux de chômage'].mean()

print(taux_chomage.info())
taux_chomage.head()
```

### Croisement des jeux de données

#### Ensemble de données groupées par total des délits (sans type de délit)

```python
# Dataframe grouped by total crimes (without type of crime)
police_records_unemployment = pd.merge(police_records_agregat.groupby(['Annee'])['Delits enregistres'].sum(),
                 taux_chomage,
                 on='Annee', 
                 how='inner')

police_records_unemployment["Taux de chômage"] = police_records_unemployment["Taux de chômage"].round(2)

police_records_unemployment
```

```python
delits = alt.Chart(police_records_unemployment).mark_line().encode(
    alt.X('Annee:O', axis=alt.Axis(title="", grid=False)),
    alt.Y('Delits enregistres:Q', axis=alt.Axis(title="Nombre de Délits enregistres", grid=False, titleColor='blue')),
    ).properties(
        title="",
        width=500,
        height=300,
    )

chomage = alt.Chart(police_records_unemployment).mark_line(color='red').encode(
    alt.X('Annee:O', axis=alt.Axis(title="", grid=False)),
    alt.Y('Taux de chômage:Q', axis=alt.Axis(title="Taux de chômage", grid=False, titleColor='red')),
    ).properties(
        title="Corrélation entre le chômage et la fréquence globale des crimes par an",
        width=500,
        height=300,
    )


alt.layer(delits, chomage).resolve_scale(
    y = 'independent'
)
```

Nous pouvons constater qu'une baisse du chômage entraîne une baisse de la criminalité, mais celle-ci n'est pas immédiate. 

En d'autres termes, nous constatons que si le chômage augmente, la criminalité augmente (2012-2015), mais apparament si le chômage diminue, la criminalité semble mettre du temps à baisser, mais elle diminue aussi. La tendance 2019-2020 confirme ce comportement. 

### Jointure des jeux de données (avec le type des délits)

```python
police_records_unemployment = pd.merge(police_records_agregat,
                 taux_chomage,
                 on='Annee', 
                 how='inner')

police_records_unemployment["Taux de chômage"] = police_records_unemployment["Taux de chômage"].round(2)

police_records_unemployment.head()
```

```python
# Save new dataframe
path_data = os.path.join(current_folder,'data\police_records_unemployment_nantes.csv')
police_records_unemployment.to_csv(path_data) 
```

### Corrélation entre le chômage et la fréquence des vols déclarés à Nantes

```python
theft_unemployment = police_records_unemployment[police_records_unemployment['Type de crime'].str.contains("Vols")]

theft_unemployment
```

```python
theft_unemployment_normalized = (theft_unemployment[["Delits enregistres","Taux de chômage"]]-theft_unemployment[["Delits enregistres","Taux de chômage"]].min())/(theft_unemployment[["Delits enregistres","Taux de chômage"]].max()-theft_unemployment[["Delits enregistres","Taux de chômage"]].min())

print(theft_unemployment_normalized.head())
```

```python
with warnings.catch_warnings(): #to Turn off Warnings
    warnings.simplefilter('ignore')
    theft_unemployment["delits_normalises"] = theft_unemployment_normalized["Delits enregistres"]
    theft_unemployment["chomage_normalise"] = theft_unemployment_normalized["Taux de chômage"]

# plot lines
plt.plot(theft_unemployment["Annee"], theft_unemployment["delits_normalises"], label = "Delits")
plt.plot(theft_unemployment["Annee"], theft_unemployment["chomage_normalise"], label = "Taux de chômage")
plt.legend()
plt.show()
```

### Correlation

```python
# Calculate the spearmans's correlation between two variables
from scipy.stats import pearsonr

corr, _ = pearsonr(theft_unemployment["delits_normalises"], theft_unemployment["chomage_normalise"])
print('Spearmans correlation: %.3f' % corr) # Correlation between unemployment and the frequency of reported theft in Nantes
```

#### Correlation entre 2018 - 2020

```python
theft_unemployment = theft_unemployment[(theft_unemployment["Annee"] >= 2018)]


with warnings.catch_warnings(): #to Turn off Warnings
    warnings.simplefilter('ignore')
    theft_unemployment["Annee"] = theft_unemployment["Annee"].astype(str)\
                        .apply(lambda x: dt.datetime.strptime(x,'%Y'))
    
    theft_unemployment["delits_normalises"] = theft_unemployment_normalized["Delits enregistres"]
    theft_unemployment["chomage_normalise"] = theft_unemployment_normalized["Taux de chômage"]

# plot lines
plt.plot(theft_unemployment["Annee"], theft_unemployment["delits_normalises"], label = "Delits")
plt.plot(theft_unemployment["Annee"], theft_unemployment["chomage_normalise"], label = "Taux de chômage")
plt.legend()
plt.figure(figsize=(15, 15))
plt.show()
```

```python
# Calculate the spearmans's correlation between two variables
corr, _ = pearsonr(theft_unemployment["delits_normalises"], theft_unemployment["chomage_normalise"])
print('Spearmans correlation: %.3f' % corr) # Correlation between unemployment and the frequency of reported theft in Nantes
```

Nous confirmons ici, mathématiquement, que ces dernières années, une baisse du chômage entraîne une baisse de la criminalité.

Il convient toutefois de mentionner que le taux de chômage n'est pas le seul indicateur de la situation de l'emploi de la population économiquement active. Une analyse parallèle peut être faite entre les demandes d'emploi enregistrées dans Pole emploi, par exemple.
